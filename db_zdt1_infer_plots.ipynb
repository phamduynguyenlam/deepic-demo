# DB-SAEA ZDT1 Inference Plots

Notebook này đọc file kết quả JSON và vẽ lại đồ thị HV cùng Pareto front.


```python
import json
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
result_path = Path(r'''C:\Users\dauma\OneDrive\Documents\HUST\Research\code\Surr-RLDE-main\db_zdt1_infer_results.json''')
result = json.loads(result_path.read_text(encoding='utf-8'))
result.keys()
```

```python
plt.figure(figsize=(8, 5))
plt.title(result['title_hv'])
plt.plot(result['db_saea']['hv_history'], marker='o', label='DB-SAEA')
plt.plot(result['nsga_eic']['hv_history'], marker='s', label='NSGA-EIC')
plt.xlabel('Step')
plt.ylabel('Hypervolume')
plt.grid(True)
plt.legend()
plt.show()
```

```python
db_front = np.asarray(result['db_saea']['final_front'], dtype=np.float32)
eic_front = np.asarray(result['nsga_eic']['final_front'], dtype=np.float32)
true_front = np.asarray(result['db_saea']['true_front'], dtype=np.float32)
plt.figure(figsize=(8, 5))
plt.title(result['title_front'])
plt.scatter(db_front[:, 0], db_front[:, 1], s=24, alpha=0.8, label='DB-SAEA')
plt.scatter(eic_front[:, 0], eic_front[:, 1], s=24, alpha=0.8, label='NSGA-EIC')
plt.plot(true_front[:, 0], true_front[:, 1], 'k-', linewidth=2, label='True Pareto Front')
plt.xlabel('f1')
plt.ylabel('f2')
plt.grid(True)
plt.legend()
plt.show()
```
